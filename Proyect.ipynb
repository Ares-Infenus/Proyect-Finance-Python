# Pregunta 2

```python
#Importando librerias
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import matplotlib_inline as mp

#Obteniendo el hmtl de la web con los datos de los ingresos totales 
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'}
req = requests.get('https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue', headers=headers)
html = BeautifulSoup(req.text, "lxml")
div = html.find_all('div', class_="col-xs-6")
div = div[0]
td = div.find_all('td')
#Limpieza y diision de fechas y Datos de Ingresos
#Calculamos la totalidad de etiquetas <td>
total_td_count = len(td)
Años = []
Revenue = []
for i in range(0,total_td_count):
    if i % 2 == 0 :
        Años.append(td[i].text)
    else:
        Revenue.append(td[i].text)

Revenue_Dataframe = pd.DataFrame()
Revenue_Dataframe['Año'] = Años
Revenue_Dataframe['Revenue'] = Revenue 
```
